```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.dates as mdate
import datetime as dt
import mplcyberpunk
```

```python
ibov = "^BVSP"
data_inicio = "2023-01-01"
data_atual = dt.datetime.now()

cotacao_ibov = yf.download(ibov, data_inicio, data_atual)["Adj Close"]

```

```python
plt.style.use("cyberpunk")

formatacao_data = mdate.DateFormatter("%b")
```

```python
fig, ax = plt.subplots()

ax.plot(cotacao_ibov.index, cotacao_ibov.values)
ax.xaxis.set_major_formatter(formatacao_data)
ax.set_title("Cotação do Ibovespa - 2023", fontweight = "bold")

plt.savefig("Ibov.png", dpi = 300)
```
