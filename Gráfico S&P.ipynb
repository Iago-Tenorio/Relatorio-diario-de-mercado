```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.dates as mdate
import datetime as dt
import mplcyberpunk
```

```python
sep = "^GSPC"
data_inicio = "2023-01-01"
data_atual = dt.datetime.now()

cotacao_sep = yf.download(sep, data_inicio, data_atual)["Adj Close"]
```

```python
plt.style.use("cyberpunk")

formatacao_data = mdate.DateFormatter("%b")
```

```python
fig, ax = plt.subplots()

ax.plot(cotacao_sep.index, cotacao_sep.values)
ax.xaxis.set_major_formatter(formatacao_data)
ax.set_title("Cotação do S&P500 - 2023", fontweight = "bold")

plt.savefig("SEP.png", dpi = 300)
```
