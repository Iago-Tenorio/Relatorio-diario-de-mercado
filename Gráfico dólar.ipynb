```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.dates as mdate
import datetime as dt
import mplcyberpunk
```

```python
dolar = "USDBRL=X"
data_inicio = "2023-01-01"
data_atual = dt.datetime.now()

cotacao_dolar = yf.download(dolar, data_inicio, data_atual)["Adj Close"]

```

```python
plt.style.use("cyberpunk")

formatacao_data = mdate.DateFormatter("%b")
```

```python
fig, ax = plt.subplots()

ax.plot(cotacao_dolar.index, cotacao_dolar.values)
ax.yaxis.set_major_formatter("R${x:1.1f}")
ax.xaxis.set_major_formatter(formatacao_data)
ax.set_title("Cotação do Dólar - 2023", fontweight = "bold")

plt.savefig("Dolar.png", dpi = 300)
```
